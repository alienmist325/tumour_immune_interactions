```python
def get_random_matrix_affinity(parameter):
    """
    Generate entries from an exponential distribution (this mimics the idea of not having most phenotypes bind). But we have renormalised and modified to keep within [0,1]
    """
    w = 1000
    h = 1000
    exp_matrix = np.random.exponential(1 / parameter, (w, h))
    print(exp_matrix)
    norm_matrix = 1 - (1 / (1 + exp_matrix))
    return norm_matrix
```

```python
values = [1/np.power(10,10), 1/np.power(10, 5), 1/10, 1, 2, 10, 100, 1000000, np.power(10,10)]

plt.rcParams['figure.figsize'] = [6, 50]

figure, axis = plt.subplots(len(values),1)

for i in range(len(values)):
    value = values[i]
    axis[i].set_xbound(0, 1000)
    print(value)
    if (value >= 1 and value <= 1000):
        axis[i].set_xlim(0,1)
    outputs = get_random_matrix_affinity(value).flatten()
    print(outputs)
    axis[i].hist(outputs)
    
```
